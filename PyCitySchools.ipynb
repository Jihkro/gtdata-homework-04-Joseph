```python
# Dependencies and Setup
import pandas as pd
import numpy as np
pd.options.mode.chained_assignment = None

# File to Load (Remember to Change These)
schooldataloc = 'schools_complete.csv'
studentdataloc = 'students_complete.csv'

# Read School and Student Data File and store into Pandas Data Frames
schooldf = pd.read_csv(schooldataloc)
studdf = pd.read_csv(studentdataloc)

# Combine the data into a single dataset
mergeddf = pd.merge(schooldf, studdf, how = "left", on="school_name")
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
totalnumofschools = len(schooldf)
totalnumofstudents = len(studdf)
totalbudget = schooldf['budget'].sum()
avgmathscore = mergeddf['math_score'].mean()
avgreadscore = mergeddf['reading_score'].mean()
totalmathpass = len(studdf.loc[studdf['math_score']>=70])
totalreadpass = len(studdf.loc[studdf['reading_score']>=70])
outputdf1 = pd.DataFrame({'Total Schools': [totalnumofschools] ,  'Total Students': [totalnumofstudents] ,
                         'Total Budget': [totalbudget] ,  'Average Math Score': [avgmathscore] ,
                         'Average Reading Score': [avgreadscore] ,  '% Passing Math': ["{:.2f}".format(100*totalmathpass/totalnumofstudents)] ,
                         '% Passing Reading': ["{:.2f}".format(100*totalreadpass/totalnumofstudents)] ,  '% Overall Passing':["{:.2f}".format(100*(totalmathpass+totalreadpass)/(2*totalnumofstudents))]})
outputdf1
```

## Top Performing Schools (By Passing Rate)

* Sort and display the top five schools in overall passing rate

```python
outputdf2 = schooldf[['school_name','type','size','budget']].rename(index = None, columns={'school_name':'School Name', 'type':'School Type','size':'Total Students', 'budget':'Total School Budget'})
outputdf2['Per Student Budget']=(outputdf2['Total School Budget']/outputdf2['Total Students'])
schoolmathtotals = []
schoolreadtotals = []
schoolmathpass = []
schoolreadpass = []
for name in outputdf2['School Name']:
    schoolmathtotal = mergeddf['math_score'].loc[(mergeddf['school_name']==name)].sum()
    schoolmathtotals.append(schoolmathtotal)
    schoolreadtotal = mergeddf['reading_score'].loc[(mergeddf['school_name']==name)].sum()
    schoolreadtotals.append(schoolreadtotal)
    schoolmathpass.append(len(mergeddf['math_score'].loc[((mergeddf['school_name']==name) & (mergeddf['math_score']>= 70))]))
    schoolreadpass.append(len(mergeddf['reading_score'].loc[((mergeddf['school_name']==name) & (mergeddf['reading_score']>= 70))]))
outputdf2['Average Reading Score'] = schoolreadtotals/outputdf2['Total Students']
outputdf2['Average Math Score'] = schoolmathtotals/outputdf2['Total Students']
outputdf2['% Passing Math'] = schoolmathpass/outputdf2['Total Students']
outputdf2['% Passing Reading'] = schoolreadpass/outputdf2['Total Students']
outputdf2['% Overall Passing Rate'] = (outputdf2['% Passing Math']+outputdf2['% Passing Reading'])/2
outputdf2.sort_values(by='% Overall Passing Rate',ascending=False).head()
```

```python
outputdf3 = outputdf2[['School Name']]
grade9 = []
grade10 = []
grade11 = []
grade12 = []
for name in outputdf3['School Name']:
    grade9.append(mergeddf['math_score'].loc[(mergeddf['school_name']==name) & (mergeddf['grade']=='9th')].mean())
    grade10.append(mergeddf['math_score'].loc[(mergeddf['school_name']==name) & (mergeddf['grade']=='10th')].mean())
    grade11.append(mergeddf['math_score'].loc[(mergeddf['school_name']==name) & (mergeddf['grade']=='11th')].mean())
    grade12.append(mergeddf['math_score'].loc[(mergeddf['school_name']==name) & (mergeddf['grade']=='12th')].mean())
    
    
outputdf3['9th'] = grade9
outputdf3['10th']=grade10
outputdf3['11th']=grade11
outputdf3['12th']=grade12
outputdf3
```
